# Matrix Fetch Operations

### Imports

```python
import numpy as np
```

### Variables

```python
a_matrix = np.matrix( [ [ 1, 2, 3 ], [ 4, 5, 6 ], [ 7, 8, 9 ] ] )
```

### Display Data

```python
print( a_matrix, "\n" )

print( a_matrix[0], "\n )
print( a_matrix[1, 0] )
print( a_matrix[2, 1:3], "\n" )

print( a_matrix.size, )
print( a_matrix.shape, "\n" )

print( np.amax( a_matrix ) )
print( np.amin( a_matrix ) )
print( np.average( a_matrix ) )
```

### Same result, slightly different syntax

```python
print( a_matrix.max() )
print( a_matrix.min() )
print( a_matrix.mean() )
```
