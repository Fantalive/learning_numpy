# Exploring numpy with a practical example of student grades

### Imports

```python
import numpy as np
```


### Variables

```python
student_grades   = np.array([ 56, 78, 98, 90, 58, 64, 67, 72, 93 ])
class_average    = np.average( student_grades )
highest_grade    = np.amax( student_grades )
lowest_grade     = np.amin( student_grades )
grade_index_hi   = np.argmax( student_grades )
grade_index_lo   = np.argmax( student_grades )
low_to_high_sort = np.sort( student_grades )
high_to_low_sort = -np.sort( -student_grades )
```


### Displaying the information from the variables


```python
print( "A list of the students grades (unordered): ", student_grades )
print( "Average of the class grades: ", class_average )
print( "The students highest grade: ", highest_grade )
print( "The students lowest grade: ", lowest_grade )
print( "Location of highest grade in list: ", grade_index_hi )
print( "Location of lowest grade in list: ", grade_index_lo )
print( "Students grades sorted from low to high: ", low_to_high_sort )
print( "Students grades sorted from high to low: ", high_to_low_sort )
```
