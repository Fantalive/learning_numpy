## Matrices

### Imports


```python
import numpy as np
```

### Variables


```python
two_by_two   = np.matrix( [[ 1, 2 ], [ 3, 4 ]] ) # Bracket required to surround both matrices [ [], [] ] like this
two_by_three = np.matrix( '1 2; 3 4; 5 6' )

zero_matrix  = np.zeros( [ 3, 2 ] )
ones_matrix  = np.zeros( [ 2, 3 ] )
```


### Printing the variables


```python
# use "\n" to space out the matrices so it's easier to read.
print( two_by_two, "\n" )
print( two_by_three, "\n" )

print( zero_matrix, "\n" )
print( ones_matrix, "\n" )
```
